# `genome_scores`

```r
genome_scores = read_csv(file = "00-Data/movie_lense/genome-scores.csv")
```

```r
str(genome_scores)
```

```r
# # of movies
length(unique(genome_scores$movieId))
```

```r
# # of tags
length(unique(genome_scores$tagId))
```

```r
# each movie has all 1128 tags 
genome_scores %>% group_by(movieId) %>% tally() %>% pull(n) %>% summary()
```

```r
13176 * 1128
```

```r
# we see the relevance is not really normalized
genome_scores %>% group_by(movieId) %>% summarise(sum_relevance = sum(relevance)) %>% 
ggplot(aes(x = sum_relevance)) + geom_histogram()
```

# oscar

```r
oscar = read_csv(file = "00-Data/the_oscar_award.csv")
str(oscar)
```

```r
unique(oscar$ceremony)
```

```r
# 92 years, 92 ceremony
unique(oscar$year_ceremony)
```

```r
tail(oscar, 10)
```

```r
# 4834 films were ever nominated
length(unique(oscar$film))
```

```r
# 1274 films ever won
length(unique(oscar$film[oscar$winner == T]))
```

```r
options(repr.plot.width=16, repr.plot.height=10)

oscar %>% group_by(year_ceremony) %>%
summarise(n_cateotry = length(unique(category)), 
          n_movies_nominated = length(unique(film)),
          n_movies_won = length(unique(film[winner]))) %>%
gather(key = 'type', value = 'n', -year_ceremony) %>%
ggplot(aes(x = year_ceremony, y = n, colour = type)) + geom_line()
```

```r
re <- "\\(([^()]+)\\)"
oscar %>% 
filter(!is.na(film)) %>%
left_join(movies %>% mutate(name = gsub('.{7}$', '', title), 
                            year = as.numeric(gsub(re, "\\1", str_extract_all(title, re)))), 
          by = c('film' = 'name', 'year_film' = 'year')) %>% 
arrange(desc(year_film)) 
```
